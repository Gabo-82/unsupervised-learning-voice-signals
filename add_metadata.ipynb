```python
import pandas as pd
import numpy as np

# Tu DataFrame principal
df = pd.read_parquet('parquets/hdbscan_results/hdbscan1000_200_umap30_9_imputed.parquet')

# Mapeo de clinical_info como arriba
clinical_info = {
    ('PF031', 'Pre'):  {'CAPE_overall': 27, 'vrqol_total': 68, 'sessions': 0},
    ('PF031', 'Post'): {'CAPE_overall': None, 'vrqol_total': 88, 'sessions': 8},
    ('PF038', 'Pre'):  {'CAPE_overall': 17, 'vrqol_total': 90, 'sessions': 0},
    ('PF038', 'Post'): {'CAPE_overall': 2,  'vrqol_total': 93, 'sessions': 5},
    ('PF111','Pre'):  {'CAPE_overall': 38, 'vrqol_total': 78, 'sessions': 0},
    ('PF111','Post'): {'CAPE_overall': None, 'vrqol_total': None, 'sessions': 11},
    ('PF126','Pre'):  {'CAPE_overall': 0,  'vrqol_total': 93, 'sessions': None},
    ('PF126','Post'): {'CAPE_overall': None, 'vrqol_total': None, 'sessions': None},
    ('PF134','Pre'):  {'CAPE_overall': 13, 'vrqol_total': 34, 'sessions': None},
    ('PF134','Post'): {'CAPE_overall': None, 'vrqol_total': None, 'sessions': None},
}

# Crear columnas vacías
df['CAPE_overall'] = np.nan
df['vrqol_total'] = np.nan
df['sessions'] = np.nan

# Agregar info por fila (basado en subject_id y week)
for (pid, week), values in clinical_info.items():
    mask = (df['subject_id'].str.upper() == pid) & (df['week'].str.lower() == week.lower())
    for col, val in values.items():
        df.loc[mask, col] = val
#guardar el DataFrame modificado
df.to_parquet('parquets/hdbscan_results/hdbscan1000_200_umap30_9_imputed.parquet', index=False)
df.head()
```
